```python
# renormalize results to get probabilities - note that my LP implementation is nto robust, due to rounding probably

# 1st: need to unpack all the Q jawns to get to 


# does this need to run for each agent or jointly?

def solve_pi(s, Q):
    # Q = (S, a1, a2) = 112 x 5 x 5
    # q_s = 5 x 5
    q = Q[s, :, :]
    
    # init vars
    a00 = cvxpy.Variable()
    a01 = cvxpy.Variable()
    a02 = cvxpy.Variable()
    a03 = cvxpy.Variable()
    a04 = cvxpy.Variable()
    a10 = cvxpy.Variable()
    a11 = cvxpy.Variable()
    a12 = cvxpy.Variable()
    a13 = cvxpy.Variable()
    a14 = cvxpy.Variable()
    a20 = cvxpy.Variable()
    a21 = cvxpy.Variable()
    a22 = cvxpy.Variable()
    a23 = cvxpy.Variable()
    a24 = cvxpy.Variable()
    a30 = cvxpy.Variable()
    a31 = cvxpy.Variable()
    a32 = cvxpy.Variable()
    a33 = cvxpy.Variable()
    a34 = cvxpy.Variable()
    a40 = cvxpy.Variable()
    a41 = cvxpy.Variable()
    a42 = cvxpy.Variable()
    a43 = cvxpy.Variable()
    a44 = cvxpy.Variable()
    
    constraints = [
        # probability constraints
        a00 >= 0,
        a01 >= 0,
        a02 >= 0,
        a03 >= 0,
        a04 >= 0,
        a10 >= 0,
        a11 >= 0,
        a12 >= 0,
        a13 >= 0,
        a14 >= 0,
        a20 >= 0,
        a21 >= 0,
        a22 >= 0,
        a23 >= 0,
        a24 >= 0,
        a30 >= 0,
        a31 >= 0,
        a32 >= 0,
        a33 >= 0,
        a34 >= 0,
        a40 >= 0,
        a41 >= 0,
        a42 >= 0,
        a43 >= 0,
        a44 >= 0,
        
        a00 + a01 + a02 + a03 + a04 + a10 + a11 + a12 + a13 + a14 + a20 + a21 + a22 + a23 + a24 + a30 + a31 + a32 + a33 + a34 + a40 + a41 + a42 + a43 + a44 == 1,
        
        #agent 1 constraints
        q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04 >= q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14,
        q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04 >= q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24,
        q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04 >= q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34,
        q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04 >= q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44,
        q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14 >= q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04,
        q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14 >= q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24,
        q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14 >= q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34,
        q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14 >= q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44,
        q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24 >= q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04,
        q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24 >= q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14,
        q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24 >= q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34,
        q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24 >= q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44,
        q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34 >= q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04,
        q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34 >= q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14,
        q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34 >= q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24,
        q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34 >= q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44,
        q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44 >= q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04,
        q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44 >= q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14,
        q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44 >= q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24,
        q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44 >= q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34,
        q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40 >= q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41,

        #agent 2 constraints
        q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40 >= q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42,
        q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40 >= q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43,
        q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40 >= q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44,
        q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41 >= q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40,
        q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41 >= q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42,
        q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41 >= q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43,
        q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41 >= q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44,
        q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42 >= q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40,
        q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42 >= q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41,
        q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42 >= q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43,
        q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42 >= q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44,
        q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43 >= q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40,
        q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43 >= q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41,
        q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43 >= q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42,
        q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43 >= q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44,
        q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44 >= q[0, 0] * a00 + q[1, 0] * a10 + q[2, 0] * a20 + q[3, 0] * a30 + q[4, 0] * a40,
        q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44 >= q[0, 1] * a01 + q[1, 1] * a11 + q[2, 1] * a21 + q[3, 1] * a31 + q[4, 1] * a41,
        q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44 >= q[0, 2] * a02 + q[1, 2] * a12 + q[2, 2] * a22 + q[3, 2] * a32 + q[4, 2] * a42,
        q[0, 4] * a04 + q[1, 4] * a14 + q[2, 4] * a24 + q[3, 4] * a34 + q[4, 4] * a44 >= q[0, 3] * a03 + q[1, 3] * a13 + q[2, 3] * a23 + q[3, 3] * a33 + q[4, 3] * a43
    ]
    
    obj = cvxpy.Maximize(q[0, 0] * a00 + q[0, 1] * a01 + q[0, 2] * a02 + q[0, 3] * a03 + q[0, 4] * a04 + q[1, 0] * a10 + q[1, 1] * a11 + q[1, 2] * a12 + q[1, 3] * a13 + q[1, 4] * a14 + q[2, 0] * a20 + q[2, 1] * a21 + q[2, 2] * a22 + q[2, 3] * a23 + q[2, 4] * a24 + q[3, 0] * a30 + q[3, 1] * a31 + q[3, 2] * a32 + q[3, 3] * a33 + q[3, 4] * a34 + q[4, 0] * a40 + q[4, 1] * a41 + q[4, 2] * a42 + q[4, 3] * a43 + q[4, 4] * a44)
    
    prob = cvxpy.Problem(obj, constraints)
    
    prob.solve()
    
    out = np.asarray([a.value for a in prob.variables()]).reshape(5, 5)
    
    out = np.maximum(out, 0)
    
    out = out / out.sum()
    
    out = out / out.sum()
    
    return out
```

```python
from copy import copy
import scipy.optimize as spo

q_delta = []

joint_q_table = np.random.random((len(all_states), 5, 5))
joint_n_table = np.zeros_like(joint_q_table)

def update_joint_q(q_table, n_table, s, a1, a2, r1, r2, s_prime, gamma=.9):
    n_table[s, a1, a2] += 1
    
#     alpha = .0001 + (1 / n_table[s, a1, a2])
    alpha = (1 / n_table[s, a1, a2])
#     alpha = .15
    
#     q_table[s, a1, a2] *= (1 - alpha)
    q_table[s, a1, a2] += alpha * ((r1 + r2) + gamma * q_table[s_prime, :, :].mean())
    return q_table, n_table


def get_actions(s, q_table):
    probs = solve_pi(s, q_table)
    
    chosen = np.random.choice(probs.flatten(), p=probs.flatten())
    res = zip(*np.where(probs == chosen)).__next__()
    return {'A': res[0], 'B': res[-1]}

for _ in tqdm(range(10000)):
    s = 31
    a_vec = get_actions(s, joint_q_table)
    r = (0, 0)

    q_old = joint_q_table[31, :, 4].mean()
    while max(r) == 0:
#     for _inner in range(100):
#         q_old = joint_q_table[31, :, 4].max()
        
        s_prime, r = sg.run_episode(a_vec['A'], a_vec['B'])
        joint_q_table, joint_n_table = update_joint_q(
            joint_q_table, joint_n_table, s, a_vec['A'], a_vec['B'], *r, s_prime
            )
    
        if np.random.rand() < .1:
            a_vec = {'A': np.random.randint(5), 'B': np.random.randint(5)}
        else:
            a_vec = get_actions(s, joint_q_table)
        s = s_prime
        
#         q_delta.append(q_old - joint_q_table[31, :, 4].max())

        if max(r) != 0:
            break

    q_delta.append(q_old - joint_q_table[31, :, 4].mean())
    sg.A = 2
    sg.B = 1
    sg.ball = 1
```
